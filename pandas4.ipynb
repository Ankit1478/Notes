# jaha bhi fair 10 euro se niche hai wha cheap aur jaha 10 se bada hai aur 20 se chota hai to mid likh do aur agar 20 se bada hai ho to high 

```python
def fare_rate(x):
    if x<10:
        return "cheap"
    elif  x>=10 and x<20:
        return "mid"
    else:
        return "high"
```

```python
df["fare_rate"]=df['Fare'].apply(fare_rate)
```
